```python
import collections
import heapq
import time
from ShortestPath import shortestpath

def main():
    file = open("C:/Users/jeetj/3D Objects/Project_2/Question_1/Undirected_Input/Input4.txt", "r")
    lines = file.readlines()
    edges = []
    n = []
        
    for line in lines:
        edges.append(line.split())
            
    last = list(edges)[-1]
    first = list(edges)[0]
    
    vertices = first[0]
    noofedges = first[1]
    g = first[2]
    
    edges.pop(0)
    if(len(last)==1):
        startnode=''.join(last)
        edges.pop()
    else:
        firstnode = list(edges)[0]
        startnode = firstnode[0]
        
    print('\nSource vertex: '+str(startnode)+'\n')
    print('Nodes: '+str(vertices)+'\n')
    print('Edges: '+str(noofedges)+'\n')
    
    if g == 'U':
        print('Undirected Graph\n')
    else:
        print('Directed Graph\n')
    
    for i in edges:
        n.append(i[0])
        
    n.extend(y[1] for y in edges)
    destnode = list((set(n)))
    
    for i in destnode:
        print ("\n"+str(startnode)+" -> "+str(i)+":")
        result = shortestpath(edges, startnode, i,g)
        print('Path Cost: '+str(result[0]))
        print('Path: '+str(result[1]))



if __name__ == "__main__":
    start_time = time.time()
    main()
    print("Runtime is %s sec" % (time.time() - start_time))
```
